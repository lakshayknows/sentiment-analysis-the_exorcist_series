```python
import re  

def remove (text):  
    pattern = re.compile('<.*?>')  
    return pattern.sub('', text)  
```

```python
def remove_url(text):  
    pattern = re.compile(r'https?://\S+|www\.\]S+')  
    return pattern.sub('', text)  
```

```python
def chat_conversion(text):  
    new_text = []  
    for w in text.split():  
        if w.upper() in sms_slang_dict:  
            new_text.append(sms_slang_dict[w.upper()])  
        else:  
            new_text.append(w)  
    return ' '.join(new_text)  
```

```python
df['Review'] = df['Review'].apply(chat_conversion)
```

```python
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):  
    new_text = []  
    
    for word in text.split():  
        if word in stopwords.words('english'):  
            new_text.append('')  
        else:  
            new_text.append(word)  
    
    x = new_text[:]  
    new_text.clear()  
    return ','.join(x)  
```

```python
df['Review'] = df['Review'].apply(remove_stopwords)
```

# Handling Emojis (If any)

```python
import re  

def remove_emoji(text):  
    emoji_pattern = re.compile(  
        u'['  
        u'\U0001F600-\U0001F64F'  # emoticons  
        u'\U0001F300-\U0001F5FF'  # symbols & pictographs  
        u'\U0001F680-\U0001F6FF'  # transport & map symbols  
        u'\U0001F1E6-\U0001F1FF'  # flags (iOS)  
        ']+', flags=re.UNICODE  
    )  
    return emoji_pattern.sub(r'', text)  
```

```python
df['Review'] = df['Review'].apply(remove_emoji)
```

```python
import gensim
import os 
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

reviews = [simple_preprocess(sentence) 
           for corpus in df['flat_reviews'] 
           for sentence in sent_tokenize(corpus)]
```

```python
len(reviews)
```

```python
model = gensim.models.Word2Vec(window=7,min_count=2)
```

```python
model.build_vocab(reviews)
```

```python
model.train(reviews, total_examples=model.corpus_count, epochs=50)  
```

```python
model.wv.most_similar('horror')  
```
