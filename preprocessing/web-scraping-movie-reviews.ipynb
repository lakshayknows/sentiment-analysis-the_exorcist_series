```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import tempfile
import time
from bs4 import BeautifulSoup
import pandas as pd

# Temporary Chrome profile to avoid session issues
temp_user_data_dir = tempfile.mkdtemp()

# Set up headless Chrome
options = Options()
options.add_argument("--headless")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument(f"--user-data-dir={temp_user_data_dir}")

# Start driver
driver = webdriver.Chrome(options=options)

# Open Rotten Tomatoes reviews page
url = 'https://www.rottentomatoes.com/tv/the_exorcist/s01/reviews?type=user'
driver.get(url)

# Load more reviews
for i in range(15):  # Adjust for more reviews
    try:
        WebDriverWait(driver, 5).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'rt-button[data-qa="load-more-btn"]'))
        )
        driver.execute_script("document.querySelector('rt-button[data-qa=\"load-more-btn\"]').click()")
        print(f"Clicked Load More {i + 1} time(s)")
        time.sleep(2)
    except Exception as e:
        print(f"No more Load More button or error at iteration {i}: {e}")
        break

# Parse loaded page
soup = BeautifulSoup(driver.page_source, 'html.parser')
review_tiles = soup.find_all('div', class_='audience-review-row')

# Extract review data
reviews_data = []

for review in review_tiles:
    # Name
    name_tag = review.find('div', attrs={'class': "audience-reviews__name-wrap"})
    name = name_tag.text.strip() if name_tag else "Anonymous"
    
    # Rating
    rating_div = review.find('div', attrs={'class': "audience-review-meta"})
    rating_tag = rating_div.find('rating-stars-group') if rating_div else None
    rating = rating_tag['score'] if rating_tag and rating_tag.has_attr('score') else "No rating"
    
    # Review Text
    review_text_tag = review.find('p', attrs={'class': 'audience-reviews__review js-review-text'})
    review_text = review_text_tag.text.strip() if review_text_tag else "No review"
    
    # Date
    date_tag = rating_div.find('span', attrs={'class': "audience-reviews__duration"}) if rating_div else None
    date = date_tag.text.strip().replace(',', '') if date_tag else "No date"

    # Append to list
    reviews_data.append({
        'Name': name,
        'Rating': rating,
        'Review': review_text,
        'Date': date
    })

# Save to DataFrame and CSV
df = pd.DataFrame(reviews_data)
df.to_csv("rottentomatoes_exorcist_detailed_reviews.csv", index=False)
print(df.head())

# Close browser
driver.quit()
```
